```python
import pandas as pd

df = pd.read_csv("/Users/moses/Desktop/Mall_Customers.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Gender'] = df['Gender'].str.strip().str.capitalize()
df['Gender'].unique()
```

```python
df.columns = (
    df.columns
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("(", "")
    .str.replace(")", "")
    .str.replace("-", "_")
)

df.columns
```

```python
df.to_csv("Mall_Customers_cleaned.csv", index=False)
```
